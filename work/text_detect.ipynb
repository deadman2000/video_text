```python
from matplotlib import pyplot as plt
import cv2
import pytesseract
import numpy as np
```

```python
image = cv2.imread('./frames/frame11.png')

plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.resize(gray, None, fx=6, fy=6)

plt.axis("off")
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
blur = cv2.GaussianBlur(gray, (9,9), 0)

plt.axis("off")
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
thresh = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11,9)

# Dilate to combine adjacent text contours
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (9,9))
dilate = cv2.dilate(thresh, kernel, iterations=4)

plt.axis("off")
plt.imshow(cv2.cvtColor(dilate, cv2.COLOR_BGR2RGB))
plt.show()
```
